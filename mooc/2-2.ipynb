1) **查询每日最高价和最低价一致的开盘记录**

```python
import pandas as pd
frame = pd.read_csv('股票数据.csv', encoding='GBK')

## 最高最低相等且非0
## print(frame[(frame['最高价'] == frame['最低价']) & (frame['最高价'])])

## 可以为0
print(frame[ frame['最高价'] == frame['最低价'] ])
```

2) **查询每日收盘价比前一交易日增长的开盘记录**

```python
import pandas as pd
frame = pd.read_csv('股票数据.csv', encoding='GBK')

print(frame[ frame['涨跌幅'] > 0 ])
```

3) **查询成交笔数大于3万或者换手率小于0.2的开盘记录**

```python
import pandas as pd
frame = pd.read_csv('股票数据.csv', encoding='GBK')

print(frame[ (frame['成交笔数'] > 30000) | (frame['换手率'] < 0.2) ])
```

4) **查询2019年12月的交易记录**

```python
import pandas as pd
frame = pd.read_csv('股票数据.csv', encoding='GBK')

print(frame[ frame['日期'].str[0:7] == '2019/12' ])
```
