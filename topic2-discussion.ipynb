1) **查询《人间词话》的借阅情况**

```python
import pandas as pd
frame = pd.read_csv('lending.dat', encoding='utf-8')

print(frame[frame['title'] == '人间词话'])
```

2) **查询 2016 年的借阅情况**

```python
import pandas as pd
frame = pd.read_csv('lending.dat', encoding='utf-8')

frame = frame[ frame['date'].str[0:4] == '2016' ]
print(frame)
```

3) **查询 1084420 用户在 2016 年的借阅情况**

```python
import pandas as pd
frame = pd.read_csv('lending.dat', encoding='utf-8')

frame = frame[ (frame['uid'] == 1084420) & (frame['date'].str[0:4] == '2016') ]
print(frame)
```

4) **查询 2016、2018、2019 年的借阅情况**

```python
import pandas as pd
frame = pd.read_csv('lending.dat', encoding='utf-8')

frame = frame[ frame['date'].str[0:4].isin(['2016', '2018', '2019']) ]
print(frame)
```

5) **查询借阅了 1103310 所借图书的用户**

```python
import pandas as pd
frame = pd.read_csv('lending.dat', encoding='utf-8')

got = frame[ frame['uid'] == 1103310 ]['bid']
frame = frame[ frame['bid'].isin(got.tolist()) ]['uid']
print(frame)
```
