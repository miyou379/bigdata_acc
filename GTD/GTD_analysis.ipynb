```python
# Subtype to Maintype mapping
# N to N mapping

__SS__ = set()

def SS_VALIDITY(f):
    @wraps(f)
    def decorated(x, cnt = 1, sub_name = ""):
        global __SS__
        f(x, cnt, sub_name)

        # Validity check after function invocation.
        return __SET_CHECK__(__SS__)
    return decorated

def __SET_CHECK__(s):
    n = len(s)
    res = dict()
    for sub, main in s:
        if sub not in res:
            res[sub] = []
        res[sub].append(main)
    return res

def __INNER_TYPE__(x, y):
    global __SS__
    __SS__.clear()

    missing_cnt = 0
    for sub, main in zip(list(GTD_data[y]), list(GTD_data[x])):
        if pd.isna(sub) and pd.isna(main):
            continue
        if pd.isna(sub):
            continue
        if pd.isna(main) == True:
            missing_cnt += 1
            continue
        if sub not in [422, 'International', 'Unknown']:
            __SS__.add((int(sub), int(main)))
            # __SS__.add((sub, main))

    if missing_cnt != 0:
        print(f"{x}: Missing main type data count: {missing_cnt}")
    return __SS__
    
@SS_VALIDITY
def __TYPE__(t, cnt = 1, sub_name = ""):
    global __SS__
    if cnt == 1:
        return __INNER_TYPE__(t, sub_name if sub_name != "" else t[:4] + 'sub' + t[4:])

    sets = []
    for i in range(1, cnt + 1):
        x = t + str(i)
        y = x[:4] + 'sub' + x[4:]
        sets.append(__INNER_TYPE__(x, y).copy())
    __SS__ = set.union(*sets)

sub_to_weap = __TYPE__('weaptype', 3)
sub_to_targ = __TYPE__('targtype', 3)
sub_to_region = __TYPE__('region', sub_name = 'country')

country_to_region = dict()

for k, v in sub_to_region.items():
    assert len(v) == 1, f"Single country matches multiple regions: {k}"
    country_to_region[k] = v[0]
```

```python
# Code to Text mapping

__SS__ = set()

def __SET_CHECK__(s):
    n = len(s)
    vis_cod, vis_txt = set(), set()
    for cod, txt in s:
        assert (cod not in vis_cod) and (txt not in vis_txt),     \
            "Duplicated/invalid <code, text> pair: ({}, {}), ({}, {})" \
            .format(cod, txt, type(cod), type(txt))
        vis_cod.add(cod)
        vis_txt.add(txt)
    return s

def __INNER_LOOP__(x):
    global __SS__
    __SS__.clear()
    y = x + "_txt"
    for ns in zip(list(GTD_data[x]), list(GTD_data[y])):
        if pd.isna(ns[0]) and pd.isna(ns[1]):
            continue
        assert pd.isna(ns[0]) == False and pd.isna(ns[1]) == False, \
            "Missing <code, text> ({}, {}) data: ({}, {})"          \
            .format(type(ns[0]), type(ns[1]), *ns)
        __SS__.add((int(ns[0]), ns[1]))
        # __SS__.add(ns)
    return __SS__

@SS_VALIDITY
def __LOOP__(t, cnt = 1, sub_name = ""):
    global __SS__
    if cnt == 1:
        return __INNER_LOOP__(t)
    
    sets = []
    for i in range(1, cnt + 1):
        x = t + str(i)
        sets.append(__INNER_LOOP__(x).copy())
    __SS__ = set.union(*sets)
        
# BEGIN_II_TERRINFO
## Only take effect when "doubtterr" == 1
## When "doubtterr" == -9, it means doubtterr variable is
## not considered when getting the data, which is invalid.
mp_alternative = dict(__LOOP__('alternative'))
# END_II_TERRINFO

# BEGIN_III_PLACE
mp_region = dict(__LOOP__('region'))
mp_country = dict(__LOOP__('country'))
# END_III_PLACE

# BEGIN_IV_ATKINFO
mp_attacktype = dict(__LOOP__('attacktype', 3))
# END_IV_ATKINFO

# BEGIN_V_WEAPINFO
mp_weaptype = dict(__LOOP__('weaptype', 4))  ## Nuclear doesn't appear in data.
mp_weaptype[10] = "Vehicle"  ## Too long!!!!
mp_weapsubtype = dict(__LOOP__('weapsubtype', 4))
# END_V_WEAPINFO

# BEGIN_VI_TARGINFO
mp_targtype = dict(__LOOP__('targtype', 3))
mp_targsubtype = dict(__LOOP__('targsubtype', 3))

tmp = dict(__LOOP__('natlty', 3))
tmp_keys = set(tmp.keys()) & set(mp_country.keys())
diff = [(k, tmp[k], mp_country[k]) for k in tmp_keys if tmp[k] != mp_country[k]]
assert len(diff) == 0, \
    "Natlty data isn't consistent with country data: {}".format(diff)
mp_natlty = { **tmp, **mp_country }
mp_country = mp_natlty
del tmp, tmp_keys, diff
# END_VI_TARGINFO

# BEGIN_VII_GINFO
mp_claimmode = dict(__SET_CHECK__(
    set.union(
        __LOOP__('claimmode').copy(),
        __LOOP__('claimmode2').copy(),
        __LOOP__('claimmode3').copy()
    )
))
# END_VII_GINFO

# BEGIN_VIII_SUM
mp_propextent = dict(__LOOP__('propextent'))
mp_hostkidoutcome = dict(__LOOP__('hostkidoutcome'))
# END_VIII_SUM

# BEGIN_IX_EX
# END_IX_EX

to_text_dict = {
    'country'       : mp_country,
    'region'        : mp_region,
    'alternative'   : mp_alternative,
    'attacktype'    : mp_attacktype,
    'targtype'      : mp_targtype,
    'targsubtype'   : mp_targsubtype,
    'natlty'        : mp_natlty,
    'claimmode'     : mp_claimmode,
    'weaptype'      : mp_weaptype,
    'weapsubtype'   : mp_weapsubtype,
    'propextent'    : mp_propextent,
    'hostkidoutcome': mp_hostkidoutcome
}

# Info of specified code
info_specificity = {
    1: "精确位置：事件发生在城市/乡村/镇",
    2: "最小的地方行政区的质心：事件发生在城市/乡村/镇，但没有可以发现的经纬度",
    3: "最小的地方行政区的质心：事件没有在城市/乡村/镇发生",
    4: "第一级行政区域的中心：没有可被识别的第二级或更小的区域，所以坐标是",
    5: "经纬度未知：没有可被识别的第一级行政区可以与恐怖袭击的地点相对应"
}

info_key = {
    'eventid': "12位的数字编码，前8位为yyyymmdd（不准确日期时eventid仍然有一个编号日期）",
    'iyear': "事件发生的起始年份",
    'imonth': "事件发生的起始月份",
    'iday': "时间发生的起始日期",
    'approxdate': "大概日期，仅当imonth或iday字段为0时有效",
    'extended': "是否持续了24个小时以上",
    'resolution': "事件的解决日期，仅当extended为1时，且事件有相关记录时有效",
    'country': "发生的国家或地区编码（根据发生时版图）",
    'region': "发生地位于世界哪个区域（共12个分类）",
    'provstate': "第一级地方行政区的名称",
    'city': "城市/村庄/城镇的名称",
    'latitude': "纬度",
    'longitude': "经度",
    'specificity': "地理编码特征，详见上方info",
    'vicinity': "是否发生在城市周边地带",
    'location': "有关事件发生地的附加位置信息",
    'summary': "事件的摘要，5W1H",
    'crit1': "是否满足入选标准1",
    'crit2': "是否满足入选标准2",
    'crit3': "是否满足入选标准3",
    'doubtterr': "疑似恐怖主义",
    'alternative': "仅当可能为恐怖主义时有效，对其分类",
    'multiple': "是否为事件组的一部分",
    'success': "袭击是否成功（一个以上攻击类型成功，暗杀除外）",
    'suicide': "是否为自杀式袭击",
    'attacktype': "攻击类型（共9个分类）",
    'targtype': "目标/受害者类型（共22个分类）",
    'targsubtype': "目标/受害者子类型（共111个分类）",
    'corp': "企业或政府机构的名称，包含未知和不适用",
    'target': "具体目标/受害者名称",
    'natlty': "具体目标/受害者国籍（飞机劫持为飞机国籍）",
    'gname': "犯罪集团名称，包含未知、个人和不适用（1998年前）",
    'gsubname': "犯罪子集团名称",
    'gname2': "超过1个时的犯罪集团名称，同gname",
    'gsubname2': "超过1个时的犯罪子集团名称，同gsubname",
    'motive': "事件的某一特定动机",
    'guncertain': "犯罪者宣称的归属是否可疑",
    'individual': "__附件2中没有提到__",
    'nperps': "凶手总数，含未报道（-99或未知）",
    'nperpcap': "抓获的凶手总数，含未报道（-99或未知）",
    'claimed': "是否有组织/个人声称对事件负责",
    'claimmode': "声称负责的方式（共10个分类）",
    'claim2': "超过1个时，是否有组织/个人声称对事件负责",
    'claimmode2': "超过1个时，声称负责的方式（共10个分类）",
    'compclaim': "是否有超过一个组织对事件单独负责（争相负责）",
    'weaptype': "武器类型（共13个分类）",
    'weapsubtype': "武器子类型（共29个分类）",
    'weapdetail': "有关武器的其他细节信息",
    'nkill': "死亡总数，含受害者和凶手",
    'nkillus': "美国公民死亡人数",
    'nkillter': "凶手死亡人数",
    'nwound': "受伤总数，含受害者和凶手",
    'nwoundus': "美国受伤人数",
    'nwoundte': "凶手受伤人数",
    'property': "是否出现财产损失，含未知（-9）",
    'propextent': "财产损害程度（共4个分类）",
    'propvalue': "财产损失价值（美元）",
    'propcomment': "财产损失相关细节",
    'ishostkid': "是否有人质或绑架，含未知（-9）",
    'nhostkid': "人质和绑架总数",
    'nhostkidus': "是否有美国公民被人质或绑架，含未知（-9）",
    'nhours': "人质和绑架事件的持续时间小时数，含未知（-99）",
    'ndays': "人质和绑架事件的持续时间天数，含未知（-99）",
    'divert': "仅在攻击类型为劫持人质（绑架）/劫持时有效，劫匪/绑匪转移到的国家",
    'kidhijcountry': "",
    'ransom': "是否索要了货币赎金，含未知（-9）",
    'ransomamt': "总索要赎金金额，含未知（-99）",
    'ransomus': "__数据中没有此列__，索要赎金的事件是否来自美国信息源，含未知（-9）",
    'ransomamtus': "来自美国信息源的索要赎金金额，含未知（-99）",
    'ransompaid': "赎金支付总额，含未知（-99）",
    'ransompaidus': "由美国来源支付的赎金金额，含未知（-99）",
    'ransomnote': "有关赎金的其他细节",
    'hostkidoutcome': "仅在攻击类型为劫持人质（绑架）/劫持时有效，人质和被绑者最终命运（共7个分类）",
    'nreleased': "释放/逃脱/获救的人质和被绑者总数",
    'addnotes': "有关攻击的附加说明",
    'scite': "关于编译具体事件信息的信息来源",
    'dbsource': "事件原始数据的来源",
    'INT_LOG': "犯罪者国籍是否与攻击地不同（国际后勤），含未知（-9）",
    'INT_IDEO': "是否攻击了与犯罪者不同国籍的目标（国际意识形态），含未知（-9）",
    'INT_MISC': "是否攻击了不同于攻击地国籍的目标（国际杂类），含未知（-9）",
    'INT_ANY': "是否是国际攻击（以上任意一种），含未知（-9）",
    'related': "仅当multiple为1时有效，记录eventid，以逗号分隔"
}
```

```python
# Function definition

def parse_eventid(s):
    assert len(s) == 12, "Eventid corrupted, length != 12"
    day, eventid = s[:8], s[8:]
    return (day, eventid)

def line_parse_crit(line_dict):
    f1, f2, f3 = [int(x) for x in [line_dict['crit1'], line_dict['crit2'], line_dict['crit3']]]
    return f1 + f2 * 2 + f3 * 4

## Judging approx by eventid is not accurate.
def line_parse_approx(line_approx):
    f_month, f_day = [x == '0' or x == 0 for x in [line_approx['imonth'], line_approx['iday']]]
    return f_month * 2 + f_day

def strip_summary_date(s):
    cur = 0
    while s[cur] != ':':
        cur += 1
    cur += 1
    if s[cur] == ' ':
        cur += 1
    return s[cur:]

def to_text(x, s):
    if pd.isna(x):
        return np.nan
    if str.isdigit(s[-1]):
        s = s[:-1]
    return to_text_dict[s][x]

def key_info(s):
    hint = "No such key, abort!"
    if s in info_key:
        hint = info_key[s]
    elif s[:-1] in info_key:
        hint = info_key[s[:-1]]
    print(hint)

def bool_check(x):
    if x in [0, 1]:
        return True
    if x == -9:
        return False
    assert False, f'Boolean check failed: {x}'

def value_check(x):
    if x >= 0:
        return True
    if x in [-9, -99, 'Unknown']:
        return False
    assert False, f'Value check failed: {x}'

# Be care that null value is not checked!
def check(x, s):
    if s in ['b', 'bool', 'boolean']:
        return bool_check(x)
    if s in ['v', 'value']:
        return value_check(x)
    if s == 'n':
        return
    assert False, f'Invalid check type {s}'

def NAN_CHECK(f):
    @wraps(f)
    def wrapper(x, *args):
        if pd.isna(x):
            return False
        return f(x, *args)
    return wrapper

nan_check = NAN_CHECK(check)
nan_bool, nan_value = NAN_CHECK(bool_check), NAN_CHECK(value_check)

def REVERSE(f):
    @wraps(f)
    def wrapper(*args, **kwargs):
        ls = f(*args, **kwargs)
        ls.reverse()
        return ls
    return wrapper

def get_colors(cnt, cm_name):
    cm = plt.get_cmap(cm_name)
    return list(cm(np.linspace(0, 1, cnt)))

reverse_colors = REVERSE(get_colors)

global_path = os.getcwd()
def within_path(f):
    @wraps(f)
    def wrapper(path, *args, **kwargs):
        global global_path
        if os.path.exists(path) == False:
            os.makedirs(path)
            print(f"{path} newly created.")
        os.chdir(path)
        f(*args, **kwargs)
        os.chdir(global_path)
    return wrapper

def reset_after_rm():
    os.chdir("../")
    os.chdir("graphs")
    print(os.getcwd())
```

## 空袭数据基本处理

```python
air_infos = {
    'airtype': 'v', 'country': 'n', 'flight_natlty':'v', 'eventid': 'v',
    'gname': 'n', 'gname2': 'n', 'gname3': 'n',
    'iyear': 'v', 'imonth': 'v', 'iday': 'v',
    'success': 'b', 'suicide': 'b', 'extended': 'b',
    'nkill': 'v', 'nwound': 'v',
    'ishostkid': 'b', 'nhostkid': 'v', 'hostkidoutcome': 'v',
    'property': 'b', 'propextent': 'v',
    'ransom': 'b', 'ransomamt': 'v'
}

def get_score(d):
    score = 100  ## base score

    if nan_value(d['nkill']):
        nkill = d['nkill']
        score += 7 * pow(nkill, 1.6)
    if nan_value(d['nwound']):
        nwound = d['nwound']
        score += 5 * pow(nwound, 1.4)

    if nan_value(d['ishostkid']):
        if d['ishostkid'] == 1:
            score += 20
            if nan_value(d['nhostkid']):
                nhostkid = d['nhostkid']
                score += 3 * pow(nhostkid, 1.2)
            if d['hostkidoutcome'] == 4:
                score += 50
    
    if nan_value(d['property']):
        if d['property'] == 1:
            score += 15
            if nan_value(d['propextent']):
                score += 100 * (4 - d['propextent'])
            
    if nan_value(d['ransom']):
        if d['ransom'] == 1:
            score += 20
            if nan_value(d['ransomamt']):
                ransomamt = d['ransomamt']
                score += 4 * ((log(ransomamt) / log(10)) ** 2)

    if d['success'] == 0:
        score *= 0.8
    if d['suicide'] == 1:
        score *= 1.5
    if d['extended'] == 1:
        score *= 1.2
    return float(score * 100 // 1 / 100)

# air = GTD[GTD['airtype'] != 0][air_infos.keys()]
```
